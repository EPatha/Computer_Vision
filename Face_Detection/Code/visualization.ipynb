```python
import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Fungsi untuk melakukan augmentasi pada gambar
def augment_image(image):
    # Augmentasi Brightness
    brightness_image = cv2.convertScaleAbs(image, alpha=1, beta=50)  # Meningkatkan kecerahan
    
    # Augmentasi Flip (horizontal flip)
    flipped_image = cv2.flip(image, 1)  # Flip horizontal
    
    # Augmentasi Grayscale
    grayscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    grayscale_image = cv2.cvtColor(grayscale_image, cv2.COLOR_GRAY2BGR)  # Mengubah kembali ke BGR
    
    # Augmentasi Shifted
    rows, cols = image.shape[:2]
    M = np.float32([[1, 0, 50], [0, 1, 50]])  # Matrix untuk pergeseran
    shifted_image = cv2.warpAffine(image, M, (cols, rows))
    
    return brightness_image, flipped_image, grayscale_image, shifted_image

# Path ke direktori val_crops
base_dir = 'data/val_crops/'

# Mengambil list semua subdirektori di dalam val_crops
subdirs = [f for f in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, f))]

# Pilih subdirektori secara acak
random_subdir = random.choice(subdirs)

# Path ke subdirektori yang dipilih
subdir_path = os.path.join(base_dir, random_subdir)

# Ambil semua gambar .ppm di dalam subdirektori
image_files = [f for f in os.listdir(subdir_path) if f.endswith('.ppm')]

# Pilih satu gambar secara acak dari subdirektori
random_image_file = random.choice(image_files)

# Path ke gambar yang dipilih
image_path = os.path.join(subdir_path, random_image_file)

# Membaca gambar asli
image = cv2.imread(image_path)

# Augmentasi gambar
brightness_image, flipped_image, grayscale_image, shifted_image = augment_image(image)

# Menampilkan gambar asli dan hasil augmentasi
fig, axes = plt.subplots(1, 5, figsize=(20, 10))

# Gambar Asli
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original')
axes[0].axis('off')

# Gambar Brightness Augmented
axes[1].imshow(cv2.cvtColor(brightness_image, cv2.COLOR_BGR2RGB))
axes[1].set_title('Brightness')
axes[1].axis('off')

# Gambar Flipped Augmented
axes[2].imshow(cv2.cvtColor(flipped_image, cv2.COLOR_BGR2RGB))
axes[2].set_title('Flipped')
axes[2].axis('off')

# Gambar Grayscale Augmented
axes[3].imshow(cv2.cvtColor(grayscale_image, cv2.COLOR_BGR2RGB))
axes[3].set_title('Grayscale')
axes[3].axis('off')

# Gambar Shifted Augmented
axes[4].imshow(cv2.cvtColor(shifted_image, cv2.COLOR_BGR2RGB))
axes[4].set_title('Shifted')
axes[4].axis('off')

# Tampilkan plot
plt.tight_layout()
plt.show()
```

```python
import os
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import tensorflow as tf
from keras.models import load_model

# Fungsi untuk mempersiapkan gambar input
def prepare_input(image):
    resized_image = tf.image.resize(image, (28, 28))  # Resize gambar ke 28x28
    grayscale_image = tf.image.rgb_to_grayscale(resized_image)  # Ubah menjadi grayscale
    normalized_image = grayscale_image / 255.0  # Normalisasi
    return tf.expand_dims(normalized_image, axis=0)  # Tambahkan dimensi batch

# Memuat model tersimpan
saved_model_path = '/home/ep/Downloads/nist_2/saved_model'
cnn_model = load_model(saved_model_path)

# Contoh data input (gunakan gambar acak)
image_prepared = prepare_input(tf.random.uniform([64, 64, 3]))  # Gambar acak 64x64

# Pastikan input berbentuk [1, 28, 28, 1] sebelum diberikan ke model
print("Prepared Input Shape:", image_prepared.shape)

# Prediksi dengan model
predictions = cnn_model(image_prepared)
print("Predictions:", predictions)

# Output tambahan jika menggunakan augmentasi
def evaluate_accuracy(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_pred = np.argmax(y_pred, axis=1)  # Ambil kelas dengan probabilitas tertinggi
    return accuracy_score(y_test, y_pred)
```
